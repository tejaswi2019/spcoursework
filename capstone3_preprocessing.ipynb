```python
stop_words = set(stopwords.words('english'))
 

result_df['paragraph_without_stopwords'] = result_df['Paragraph'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
result_df['Q1_without_stopwords'] = result_df['Q1'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
result_df['Q2_without_stopwords'] = result_df['Q2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
result_df['Q3_without_stopwords'] = result_df['Q3'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
result_df['Q4_without_stopwords'] = result_df['Q4'].apply(lambda x: ' '.join([word for word in str(x).split() if word not in (stop_words)]))
#print(word_tokens)
result_df.head()
#print(filtered_sentence)
# remove one and multiple
# isolate question and answers
# extract question new column
#Q1_A,Q1_B,Q1_C,Q1_D separately
#lemmatize or stemming paragraph, question and answers(4)

```

```python
result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace('*', '')
result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace('?', '')
#result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace(':', '')
result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace('multiple:', '')
result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace('one:', '')
result_df['Q1_without_stopwords']= result_df['Q1_without_stopwords'].str.replace(':', '')
result_df['Q1_without_stopwords'] = result_df['Q1_without_stopwords'].astype(str)
#final_df.head()
result_df.head()
#result_df.dtypes
```

```python
df1_Q1 = result_df['Q1_without_stopwords'].apply(lambda x: pd.Series(str(x).split('A)')))
df1_Q1.rename(columns={0: 'Q1',1:'rest'}, inplace=True)
df1_Q1.head()
```

```python
df1_Q1_A = df1_Q1['rest'].apply(lambda x: pd.Series(str(x).split('B)')))
df1_Q1_A.rename(columns={0: 'Q1_A',1:'rest'}, inplace=True)
df1_Q1_A.head()
```

```python
df1_Q1_B = df1_Q1_A['rest'].apply(lambda x: pd.Series(str(x).split('C)')))
df1_Q1_B.rename(columns={0: 'Q1_B',1:'rest'}, inplace=True)
df1_Q1_B.head()
```

```python
df1_Q1_C = df1_Q1_B['rest'].apply(lambda x: pd.Series(str(x).split('D)')))
df1_Q1_C.rename(columns={0: 'Q1_C',1:'Q1_D'}, inplace=True)
df1_Q1_C.head()
```
