```python
# Get cosine Similarity
def compute_cosine(input_1, input_2):

  input_1 = input_1#.iloc[0,:]
  input_2 = input_2.iloc[0,:]

  return(np.dot(input_1,input_2)/(norm(input_1)*norm(input_2)))

def get_closest_match_from_Q(loc_para, loc_Q):
  loc_para_split = loc_para.split('. ')

  # TFIDF Section
  
  # Paragraph Transformer
  vectorizer = TfidfVectorizer(smooth_idf=True,use_idf=True)
  vect_result = vectorizer.fit_transform(loc_para_split)
  loc_para_tfidf = pd.DataFrame(vect_result.toarray(), columns=vectorizer.get_feature_names())
  
  # Question Transformer
  vect_result = vectorizer.transform([loc_Q])
  loc_Q_tfidf = pd.DataFrame(vect_result.toarray(), columns=vectorizer.get_feature_names())

  cosine_results = loc_para_tfidf.apply(lambda loc_tfidf_row: compute_cosine(loc_tfidf_row, loc_Q_tfidf), axis = 1)
  final_ans_full = loc_para_split[cosine_results[cosine_results == max(cosine_results)].index[0]]
  return(final_ans_full)
```
